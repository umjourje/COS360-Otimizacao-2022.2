```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
%matplotlib inline
```

```python
def f(x, y):
    return ((x**2) + (np.sin(x+y)) + (y**4))

def gradient(x, y):
    return np.array((2*x + math.cos(x+y) , 4*y**3 + math.cos(x+y) ),dtype='double').reshape(2,1)

def hessian(x, y):
    return np.array(([2-math.sin(x+y), -math.sin(x+y)],
                     [-math.sin(x+y) , 12*(y**2) -math.sin(x+y)]),dtype='double').reshape(2,2)
```

```python
n_amostras = 40
x = np.linspace(-5, 5, n_amostras)
y = np.linspace(-5, 5, n_amostras)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
def f1(x,y):
    return x**2

def f2(x,y):
    return np.sin(x+y)

def f3(x,y):
    return y**4

def f1_f3(x,y):
    return (x**2)+(y**4)

def f1_f2(x,y):
    return (x**2)+np.sin(x+y)

def f3_f2(x,y):
    return (y**4)+np.sin(x+y)
```

```python
# funções simples
f1_plot = f1(X, Y)
f2_plot = f2(X, Y)
f3_plot = f3(X, Y)

# funções compostas
f1_f3_plot = f1_f3(X, Y)
f1_f2_plot = f1_f2(X, Y)
f3_f2_plot = f3_f2(X, Y)
```

```python
ax_f1 = plt.axes(projection='3d')
ax_f1.plot_surface(X, Y, f1_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax_f1.set_title('3D Surface for x^2');
```

```python
ax_f2 = plt.axes(projection='3d')
ax_f2.plot_surface(X, Y, f2_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax_f2.set_title('3D Surface for sin(x+y)');
```

```python
ax_f3 = plt.axes(projection='3d')
ax_f3.plot_surface(X, Y, f3_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax_f3.set_title('3D Surface for y^4');
```

```python
ax_f1_f3 = plt.axes(projection='3d')
ax_f1_f3.plot_surface(X, Y, f1_f3_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax_f1_f3.set_title('Surface for x^2+y^4');
```

```python
ax_f1_f2 = plt.axes(projection='3d')
ax_f1_f2.plot_surface(X, Y, f1_f2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax_f1_f2.set_title('Surface for x^2 + sin(x+y)');
```

```python
# somando os resultados de f3 e f2
f3_f2 = f3_plot+f2_plot

ax_f3_f2 = plt.axes(projection='3d')
ax_f3_f2.plot_surface(X, Y, f3_f2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax_f3_f2.set_title('Surface for y^4 + sin(x+y)');
```

```python
fig2 = plt.figure()
sns.set_style('whitegrid')
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('f(x,y)')
ax.set_xlim3d([-5, 5])
ax.set_ylim3d([-5, 5])
ax.set_zlim3d([-1, 5])
plt.show()
```
